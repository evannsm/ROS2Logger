```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Font and figure setup
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "Helvetica",
    "font.size": 14
})

# === CONFIGURATION ===
data_path = 'z_hardware_evanns_logs/'
line_thickness = 5  # <-- Change this to adjust all plotted line widths

# File dictionaries
FBL_dict = {
    '6': 'official_FBL_HX.log',
    '7': 'official_FBL_CH_SPN.log',
    '8': 'official_FBL_HX_SPN.log',
}

MPC_dict = {
    '6': 'official_NMPC_HX.log',
    '7': 'official_NMPC_CH_SPN3.log',
    '8': 'official_NMPC_HX_SPN.log',
}

NR_dict = {
    '6': 'official_NR_HX.log',
    '7': 'official_NR_CH_SPN3.log',
    '8': 'official_NR_HX_SPN.log',
}

controllers = {'NR': NR_dict, 'NMPC': MPC_dict, 'FBL': FBL_dict}

# Titles for each subplot
title_labels = [
    'a) NR: Helix A – Y vs X', 'b) NR: Helix A – Z vs X',
    'c) NR: Circle C – Y vs X', 'd) NR: Circle C – Yaw vs Time',
    'e) NR: Helix B – Y vs X', 'f) NR: Helix B – Z vs X', 'g) NR: Helix B – Yaw vs Time',

    'h) NMPC: Helix A – Y vs X', 'i) NMPC: Helix A – Z vs X',
    'j) NMPC: Circle C – Y vs X', 'k) NMPC: Circle C – Yaw vs Time',
    'l) NMPC: Helix B – Y vs X', 'm) NMPC: Helix B – Z vs X', 'n) NMPC: Helix B – Yaw vs Time',

    'o) FBL: Helix A – Y vs X', 'p) FBL: Helix A – Z vs X',
    'q) FBL: Circle C – Y vs X', 'r) FBL: Circle C – Yaw vs Time',
    's) FBL: Helix B – Y vs X', 't) FBL: Helix B – Z vs X', 'u) FBL: Helix B – Yaw vs Time',
]

fig, axs = plt.subplots(3, 7, figsize=(52, 25), sharex=False, sharey=False)
axs = axs.flatten()

fw = 'medium'
fs_title = 36
fs_label = 29
tick_fontsize = 14
margin = 0.15

for i, (ctrl_idx, ctrl_name) in enumerate(zip([0, 1, 2], ['NR', 'NMPC', 'FBL'])):
    ctrl_dict = controllers[ctrl_name]

    # Helix A
    df = pd.read_csv(os.path.join(data_path, ctrl_dict['6']))
    for k, (ycol, label, subplot_offset) in enumerate([('y', 'y (m)', 0), ('z', 'z (m)', 1)]):
        ax = axs[i * 7 + subplot_offset]
        ax.plot(df['x'], df[ycol], 'r-', label='Actual', linewidth=line_thickness)
        ax.plot(df['x_ref'], df[f'{ycol}_ref'], 'b--', label='Reference', linewidth=line_thickness)
        ax.set_xlabel('x (m)', fontsize=fs_label, fontweight=fw)
        ax.set_ylabel(label, fontsize=fs_label, fontweight=fw)
        ax.set_title(title_labels[i * 7 + subplot_offset], fontsize=fs_title, fontweight=fw)
        ax.tick_params(labelsize=tick_fontsize)
        ax.set_aspect('equal', adjustable='datalim')

    # Circle C
    df = pd.read_csv(os.path.join(data_path, ctrl_dict['7']))
    axs[i * 7 + 2].plot(df['x'], df['y'], 'r-', label='Actual', linewidth=line_thickness)
    axs[i * 7 + 2].plot(df['x_ref'], df['y_ref'], 'b--', label='Reference', linewidth=line_thickness)
    axs[i * 7 + 2].set_xlabel('x (m)', fontsize=fs_label, fontweight=fw)
    axs[i * 7 + 2].set_ylabel('y (m)', fontsize=fs_label, fontweight=fw)
    axs[i * 7 + 2].set_title(title_labels[i * 7 + 2], fontsize=fs_title, fontweight=fw)
    axs[i * 7 + 2].tick_params(labelsize=tick_fontsize)
    axs[i * 7 + 2].set_aspect('equal', adjustable='datalim')

    axs[i * 7 + 3].plot(df['time'], df['psi'], 'r-', label='Actual', linewidth=line_thickness)
    axs[i * 7 + 3].plot(df['time'], df['psi_ref'], 'b--', label='Reference', linewidth=line_thickness)
    axs[i * 7 + 3].set_xlabel('time (s)', fontsize=fs_label, fontweight=fw)
    axs[i * 7 + 3].set_ylabel('yaw (rad)', fontsize=fs_label, fontweight=fw)
    axs[i * 7 + 3].set_title(title_labels[i * 7 + 3], fontsize=fs_title, fontweight=fw)
    axs[i * 7 + 3].tick_params(labelsize=tick_fontsize)

    # Helix B
    df = pd.read_csv(os.path.join(data_path, ctrl_dict['8']))
    for k, (ycol, label, offset) in enumerate([
        ('y', 'y (m)', 4), ('z', 'z (m)', 5)
    ]):
        ax = axs[i * 7 + offset]
        ax.plot(df['x'], df[ycol], 'r-', label='Actual', linewidth=line_thickness)
        ax.plot(df['x_ref'], df[f'{ycol}_ref'], 'b--', label='Reference', linewidth=line_thickness)
        ax.set_xlabel('x (m)', fontsize=fs_label, fontweight=fw)
        ax.set_ylabel(label, fontsize=fs_label, fontweight=fw)
        ax.set_title(title_labels[i * 7 + offset], fontsize=fs_title, fontweight=fw)
        ax.tick_params(labelsize=tick_fontsize)
        ax.set_aspect('equal', adjustable='datalim')

    axs[i * 7 + 6].plot(df['time'], df['psi'], 'r-', label='Actual', linewidth=line_thickness)
    axs[i * 7 + 6].plot(df['time'], df['psi_ref'], 'b--', label='Reference', linewidth=line_thickness)
    axs[i * 7 + 6].set_xlabel('time (s)', fontsize=fs_label, fontweight=fw)
    axs[i * 7 + 6].set_ylabel('yaw (rad)', fontsize=fs_label, fontweight=fw)
    axs[i * 7 + 6].set_title(title_labels[i * 7 + 6], fontsize=fs_title, fontweight=fw)
    axs[i * 7 + 6].tick_params(labelsize=tick_fontsize)

plt.tight_layout(rect=[0, 0.03, 1, 1])
plt.savefig("trajectories_comparison_3x7_spin.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
